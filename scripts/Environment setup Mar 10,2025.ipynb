```python
#Step 20 (Optional)

import tensorflow as tf

import time

# Create a simple computation

def test_tensorflow_on_gpu():

    print("Testing GPU computation...")

    # Set TensorFlow to use GPU

    if tf.config.list_physical_devices('GPU'):

        tf.config.experimental.set_memory_growth(tf.config.list_physical_devices('GPU')[0], True)

    # Generate a large random matrix

    shape = (1000, 1000)

    a = tf.random.uniform(shape)

    b = tf.random.uniform(shape)

    # Run matrix multiplication on GPU and measure time

    start = time.time()

    with tf.device('/GPU:0'):  # Force computation on GPU

        c = tf.matmul(a, b)

    gpu_time = time.time() - start

    print(f"‚úÖ GPU Computation Time: {gpu_time:.4f} seconds")

    # Run the same on CPU for comparison

    start = time.time()

    with tf.device('/CPU:0'):  # Force computation on CPU

        c_cpu = tf.matmul(a, b)

    cpu_time = time.time() - start

    print(f"‚úÖ CPU Computation Time: {cpu_time:.4f} seconds")

    if gpu_time < cpu_time:

        print("üéâ GPU is working and accelerating computations!")

    else:

        print("‚ö†Ô∏è GPU might not be properly used.")

test_tensorflow_on_gpu()

```
